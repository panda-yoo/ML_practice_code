```julia
function prime_factors(n::Int)
    result = []
    i = 2
    while n > 1
        while n % i == 0
            push!(result, i)
            (n /= i)
        end
        (i += 1)
    end
    return result
end

# Usage: prime_factors(number)
```

```julia
prime_factors(7398)
```
